```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Ignore any warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load 
mouse_drug_data = "Data/mouse_drug_data.csv"
clinical_trial_data = "Data/clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Dataset and the Clinical Trial Dataset to Create New Dataset mouse_drug
mouse_data = pd.read_csv(mouse_drug_data)
clinical_data = pd.read_csv(clinical_trial_data) 
clinical_data
```

```python
mouse_data
```

```python
# Mouse data has two drugs that are not part of analysis, so they will be removed from dataset
# (Source working with J. Hixson)
clinical_data = clinical_data.drop(clinical_data.loc[clinical_data['Mouse ID']=='g989'].index)
mouse_data = mouse_data.drop(mouse_data.loc[mouse_data['Mouse ID']=='g989'].index)
```

```python
# Combine the dataset into a single dataset
combined_data_df = pd.merge(clinical_data, mouse_data, how='outer', on='Mouse ID')

# Reduce data to perform analysis ONLY for Capomulin, Infubinol, Ketapril, and Placebo
drugs = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']
reduced_data_df = combined_data_df.loc[combined_data_df['Drug'].isin(drugs)]
reduced_data_df
```

## Tumor Response to Treatment

```python
# Compute Mean for each Tumor Volume Data in group
tumor_mean = reduced_data_df.groupby(['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)']

# Convert to DataFrame
tumor_mean_df = pd.DataFrame(tumor_mean)
```

```python
# Preview DataFrame
tumor_mean_df
```

```python
# Separate SEM value of the Tumor Volumes Grouped by Drug and Timepoint
# (Source working with J. Hixson)
tumor_error = reduced_data_df.groupby(['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)']

# Convert to DataFrame
tumor_error_df = pd.DataFrame(tumor_error)

# Preview DataFrame
tumor_error_df
```

```python
# Re Format Data Frames by Indexes by Timepoint, Column = Drugs and mean of tumor size
mean_reformat = tumor_mean_df.reset_index()
mean_pivot = mean_reformat.pivot(index='Timepoint', columns='Drug')['Tumor Volume (mm3)']
error_reformat = tumor_error_df.reset_index()
error_pivot = error_reformat.pivot(index='Timepoint', columns='Drug')['Tumor Volume (mm3)']

# Preview data sets
mean_pivot
```

```python
#Set up Scatterplot
plt.figure(figsize=(10,5))

# Plot treatment by days using index for x-axis, mean for y-axis and error bars for deviation of y
plt.errorbar(mean_pivot.index, mean_pivot["Capomulin"], yerr=error_pivot["Capomulin"],\
             color="y", marker="^", markersize=8, linestyle="dotted", linewidth=1.00)
plt.errorbar(mean_pivot.index, mean_pivot["Infubinol"], yerr=error_pivot["Infubinol"],\
             color="g", marker="D", markersize=8, linestyle="dotted", linewidth=1.00)
plt.errorbar(mean_pivot.index, mean_pivot["Ketapril"], yerr=error_pivot["Ketapril"],\
             color="k", marker="v", markersize=8, linestyle="dotted", linewidth=1.00)
plt.errorbar(mean_pivot.index, mean_pivot["Placebo"], yerr=error_pivot["Placebo"],\
             color="b", marker="s", markersize=8, linestyle="dotted", linewidth=1.00)

# Set x and y axis labels and title of the chart
plt.title("Tumor Treatment Outcomes", fontsize=20, fontweight='bold')
plt.xlabel("Days in Study", fontsize=16, fontweight='bold') 
plt.ylabel("Tumor Volume (mm3)", fontsize=16, fontweight='bold') 

# Place dashed grid and legend in best location
plt.grid(linestyle="dotted")
plt.legend(drugs, loc='best', fontsize=14, fancybox=True)

# Show the plot
plt.show()
```

```python
# Separate Mean of Metastatic sites grouped by Drug and Timepoint
meta_mean = reduced_data_df.groupby(['Drug', 'Timepoint']).mean()['Metastatic Sites']

# Convert to DataFrame
meta_mean_df = pd.DataFrame(meta_mean)

# Preview DataFrame
meta_mean_df.head()
```

```python
# Separate the SEM values of Metastatic Sites Grouped by Drug and Timepoint
# (Source working with J. Hixson)
meta_error = reduced_data_df.groupby(['Drug', 'Timepoint']).sem()['Metastatic Sites']
# Convert to DataFrame
meta_error_df = pd.DataFrame(meta_error)
# Preview DataFrame
meta_error_df
```

## Metastatic Response to Treatment

```python
# Re Format the Data Frames, Indexes by timepoint, columns by drug name and tumor mean and error values by df
# (Source working with J. Hixson)
meta_reformat = meta_mean_df.reset_index()
meta_pivot = meta_reformat.pivot(index='Timepoint', columns='Drug')['Metastatic Sites']
meta_error_reformat = meta_error_df.reset_index()
meta_error_pivot = meta_error_reformat.pivot(index='Timepoint', columns='Drug')['Metastatic Sites']

# Preview data sets
meta_error_pivot.head()
```

```python
#Set up Scatterplot
plt.figure(figsize=(10,5))

# Plot treatment by days using index for x-axis, mean for y-axis and error bars for deviation of y

plt.errorbar(mean_pivot.index, mean_pivot["Capomulin"], yerr=error_pivot["Capomulin"],\
             color="y", marker="^", markersize=8, linestyle="dotted", linewidth=1.00)
plt.errorbar(mean_pivot.index, mean_pivot["Infubinol"], yerr=error_pivot["Infubinol"],\
             color="g", marker="D", markersize=8, linestyle="dotted", linewidth=1.00)
plt.errorbar(mean_pivot.index, mean_pivot["Ketapril"], yerr=error_pivot["Ketapril"],\
             color="k", marker="v", markersize=8, linestyle="dotted", linewidth=1.00)
plt.errorbar(mean_pivot.index, mean_pivot["Placebo"], yerr=error_pivot["Placebo"],\
             color="b", marker="s", markersize=8, linestyle="dotted", linewidth=1.00)

# Set x and y axis labels and title of the chart
plt.title("Metastatic Spread During Treatment", fontsize=20, fontweight='bold')
plt.xlabel("Days of Study", fontsize=16, fontweight='bold') 
plt.ylabel("Metastatic Sites", fontsize=16, fontweight='bold') 

# Place dashed grid and legend in best location
plt.grid(linestyle="dashed")
plt.legend(drugs, loc='best', fontsize=14, fancybox=True)

# Show the plot
plt.show()
```

## Survival Rates

```python
# Compute the count of mice grouped by Drug and Timepoint
mice_count = reduced_data_df.groupby(['Drug', 'Timepoint']).count()['Mouse ID']

# Convert to DataFrame
mice_count_df = pd.DataFrame({"Mouse Count": mice_count})

# Preview DataFrame
mice_count_df
```

```python
# Re Format Data Frames of mice count
mice_reformat = mice_count_df.reset_index()
mice_pivot = mice_reformat.pivot(index='Timepoint', columns='Drug')['Mouse Count']

# Preview Data Frame
mice_pivot.head()
```

```python
# Set up splatterplot
plt.figure(figsize=(10,6))

# Create the Plot (Accounting for percentages) dividing by number of mice to start with 
# (Source https://stackoverflow.com/questions/22239691/code-for-best-fit-straight-line-of-a-scatter-plot-in-python)
plt.plot(100 * mice_pivot["Capomulin"] / 25, "bv", linestyle="-", markersize=10, linewidth=1.00)
plt.plot(100 * mice_pivot["Infubinol"] / 25, "ms", linestyle="-", markersize=10, linewidth=1.00)
plt.plot(100 * mice_pivot["Ketapril"] / 25, "g", linestyle="-", markersize=10, linewidth=1.00)
plt.plot(100 * mice_pivot["Placebo"] / 25 , "yD", linestyle="-", markersize=10, linewidth=1.00)

# Set x and y axis labels and title for the chart
plt.title("Survival Outcomes During Treatment", fontsize=20, fontweight='bold')
plt.ylabel("Survival Rate (%)",  fontsize=16, fontweight='bold') 
plt.xlabel("Days in Study", fontsize=16, fontweight='bold') 
plt.grid(True)
plt.legend(drugs, loc="best", fontsize=14, fancybox=True)

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Change in tumor size over 45 days for individual drug
# (Source working with J. Hixson)
capchange = ((mean_pivot["Capomulin"][45] - mean_pivot["Capomulin"][0])/45)*100
infuchange = ((mean_pivot["Infubinol"][45] - mean_pivot["Infubinol"][0])/45)*100
ketachange = ((mean_pivot["Ketapril"][45] - mean_pivot["Ketapril"][0])/45)*100
placchange = ((mean_pivot["Placebo"][45] - mean_pivot["Placebo"][0])/45)*100
```

```python
# Create new dataframe
drug_change_df = pd.DataFrame({"Drug": ["Capomulin", "Infubinol", "Ketapril", "Placebo"],
                               "Percent Change": [capchange, infuchange, ketachange, placchange]
                              })
drug_change_df
```

```python
# PLot Data
plt.figure(figsize=(10, 5))
plt.xlim(-1, 4)
plt.ylim(-20, 80)
plt.grid(True)
plt.title("Tumor Outcome Percent Change In 45 Day Treatment", fontsize=16)
plt.xlabel("Drug Name", fontsize=12)
plt.ylabel("% Tumor Volume Change", fontsize=12)
count = 0
for x in drug_change_df['Percent Change']:
   if x < 0:
       y_value = -5
   else:
       y_value = 2
   plt.text(count, y_value, str(round(x, 2)) + '%', ha = 'center', color = 'white', fontsize = 16)
   count += 1
plt.bar("Drug", "Percent Change", data=drug_change_df, color=['r','b','b','b'])
plt.show()
```
